```python
import numpy as np 
import pandas as pd 
from autogluon.tabular import TabularDataset, TabularPredictor
import os
```

```python
train = pd.read_csv('/kaggle/input/vk-test/train_df.csv')
test = pd.read_csv('/kaggle/input/vk-test/test_df.csv')
```

```python
x_train = TabularDataset(train)
x_test = TabularDataset(test)
label = 'target'
time_limit = 25200
```

```python
metric = 'balanced_accuracy'
predictor = TabularPredictor(label, eval_metric=metric).fit(x_train, presets='best_quality', num_bag_folds=5, num_stack_levels=1)
```

```python
predictor.evaluate(x_test, silent=True)
```

```python
predictor.leaderboard(x_test, extra_metrics=['roc_auc', 'log_loss', 'accuracy', 'f1', 'f1_weighted']).head(30)
```
